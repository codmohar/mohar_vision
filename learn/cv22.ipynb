CROPPING IMAGE

```python

cv_lam_bgr=cv2.imread("car.jpeg",cv2.IMREAD_COLOR)
plt.imshow(cv_lam_bgr)
car=cv_lam_bgr[:,:,::-1]
plt.imshow(car)
```

```python
cropped_car=car[400:1100,200:2300]    #y1:y2,x1:x2
plt.imshow(cropped_car)
```

Resizing Images

```python
resized_car=cv2.resize(cropped_car,None,fx=2,fy=2)
plt.imshow(resized_car)
```

```python
# --- IGNORE ---
#Without aspect resizing
x_width=1000
y_width=2000

dim=(x_width,y_width)
resized_car=cv2.resize(cropped_car,dsize=dim,interpolation=cv2.INTER_AREA)
plt.imshow(resized_car)
```

Resizing while maintaining aspect ratio

```python
#using aspect ratio/desize
desired_width=100
aspect_ratio=desired_width/cropped_car.shape[1]
desired_height =int(cropped_car.shape[0]*aspect_ratio)
dim=(desired_width,desired_height)

resized_car=cv2.resize(cropped_car,dsize=dim,interpolation=cv2.INTER_AREA)
plt.imshow(resized_car)
```
