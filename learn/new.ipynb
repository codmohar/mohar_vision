```python
img_man=cv2.imread("man.JPG",cv2.IMREAD_COLOR)  #1 for color
print(img_man)  #numpy array
print(img_man.shape)  #height, width, channels
```

```python
b,g,r=cv2.split(img_man)
print(b)
print(g)    
print(r)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1);plt.imshow(r, cmap='gray');plt.title("Red Channel")
plt.subplot(1,3,2);plt.imshow(g, cmap='gray');plt.title("Green Channel")
plt.subplot(1,3,3);plt.imshow(b, cmap='gray');plt.title("Blue Channel")


#cobined rgb channels
```

```python
merged_img=cv2.merge((b,g,r))
plt.imshow(merged_img[:,:,::-1]);plt.title("Merged Image")
```

Converting Two different color spaces cv2.cvtColor() 

That is equal to cv2.cvtColor(src, code). That is the output image of the same size and depth as src, the input image (8-bit unsigned, 16-bit unsigned, or single precision floating point). Color space conversion code 

```python
real_img=cv2.cvtColor(merged_img, cv2.COLOR_BGR2RGB)
plt.imshow(real_img);plt.title("Real Image")
```

```python
img_hsv=cv2.cvtColor(merged_img, cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(img_hsv)
```

```python
h_new=h+10
img_new_merged=cv2.merge((h_new,s,v))
img_new_rgb=cv2.cvtColor(img_new_merged, cv2.COLOR_HSV2RGB)
plt.imshow(img_new_rgb);plt.title("New Image with H+10")

plt.figure(figsize=(10,5))
plt.subplot(1,3,1);plt.imshow(h, cmap='gray');plt.title("H Image")
plt.subplot(1,3,2);plt.imshow(s, cmap='gray');plt.title("S Image")
plt.subplot(1,3,3);plt.imshow(v, cmap='gray');plt.title("V Image")
```
