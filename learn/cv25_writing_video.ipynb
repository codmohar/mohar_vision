Read Video from Source

```python
#importing library

import cv2
import matplotlib.pyplot as plt
#%matplotlib_inline

source="car_racing.mp4"  # source =0 for webcam

cap=cv2.VideoCapture(source)
```

```python
if (cap.isOpened()==False):
    print("Error opening video stream or file")
```

Read and Display one frame

```python
ret, frame = cap.read()
if ret==True:
    print(frame.shape)
    plt.imshow(frame[:,:,::-1])
```

```python
#writting video using opencv
#default revolution of the brain
#Videowriter object =cv.Videowriter(filename,fourcc,fps,frame_size,isColor)

frame_width=int(cap.get(3))
frame_height=int(cap.get(4))

out_avi=cv2.VideoWriter('car_racing.avi',cv2.VideoWriter_fourcc('M','J','P','G'),10,(frame_width,frame_height))
out_mp4=cv2.VideoWriter('car_racing.mp4',cv2.VideoWriter_fourcc(*'XVID'),10,(frame_width,frame_height))
```

```python
TODO:"Read frames and write into output files"

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        out_avi.write(frame)
        out_mp4.write(frame)
    else:
        break
    
    
cap.release()
out_avi.release()   
out_mp4.release()
```
