Annotating Images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = [0.9, 0.9]
from IPython.display import Image
```

```python
image=cv2.imread("rocket.png",cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap="gray")
```

```python

cv_lam_bgr=cv2.imread("car.jpeg",cv2.IMREAD_COLOR)
plt.imshow(cv_lam_bgr)
car=cv_lam_bgr[:,:,::-1]
plt.imshow(car)
```

Drawing line

```python
from cv2 import LINE_AA


cv2.line(image,(200,400),(100,600),(0,255,255),5,lineType=LINE_AA)
plt.imshow(image)
```

Drawing Circle

```python

image_circle=image.copy()
cv2.circle(image_circle,(500,400),300,(255,0,0),5,lineType=LINE_AA)
plt.imshow(image_circle)
```

```python
# Draw a rectangle
image_rectangle=image.copy() 
cv2.rectangle(image_rectangle,(300,0),(700,500),(0,255,0),5)
plt.imshow(image_rectangle)
```

```python
#Adding text
image_text=image.copy()
cv2.putText(image_text,"OpenCV",(300,400),cv2.FONT_HERSHEY_SIMPLEX,3,(255,255,0),5)
plt.imshow(image_text)

```
